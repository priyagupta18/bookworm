### Popularity based recommendation system 

I will extract those TOp 50 books which has more than 250 rating

```python
ratings_with_name = rating_df.merge(book_df, on = 'ISBN')
ratings_with_name.head() 
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].count().reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
```

```python
popular_df.head()
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df = popular_df.merge(book_df, on ='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
popular_df.reset_index(drop=True)
```

### Collaborative recommendation system 

I will consider only users with >200 ratings and books with >50 ratings

```python
x = ratings_with_name.groupby('User-ID')["Book-Rating"].count()>200
target_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(target_users)]
filtered_rating.head()
```

```python
y = filtered_rating.groupby('Book-Title')['Book-Rating'].count()>=50
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings.head()
```

```python
final_ratings.shape
```

```python
final_ratings['Book-Title'].nunique()
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0, inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
similarity_score
```

```python
book_index = list(pt.index.values)

def recommend(book_name):
    b_index = book_index.index(book_name)
    recs = []
    dis = similarity_score[b_index]
    book_list = sorted(list(enumerate(dis)), reverse=True, key = lambda x:x[1])[1:6]
    for i, book in book_list:
        recs.append(book_index[i])
    return recs
    

recommend("1984")
```

```python
#extracting images from final_ratings dataframe
y = recommend("1984")
for rec in y:
    print(final_ratings[final_ratings['Book-Title']==rec]['Image-URL-M'].values[0])
```
